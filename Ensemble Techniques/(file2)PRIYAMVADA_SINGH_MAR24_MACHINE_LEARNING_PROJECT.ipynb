```python
import re
import string
from nltk.stem.snowball import SnowballStemmer
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
all_speeches=inaugural.raw(('1941-Roosevelt.txt','1961-Kennedy.txt','1973-Nixon.txt'))
```

```python
stopwords=nltk.corpus.stopwords.words("english")+list(string.punctuation)+["mr.","mrs.","'s"]
```

```python
stemmer=SnowballStemmer("english")
```

```python
def myTokenizer(text):
    temp_tokens=[word for word in nltk.word_tokenize(text)]
    tokens=[]
    for token in temp_tokens:
        if re.search('[a-zA-Z]',token):
            tokens.append(token)
    ret_tokens=[stemmer.stem(t) for t in tokens]
    return ret_tokens
```

```python
tokens_roo=myTokenizer(speech1)
tokens_ken=myTokenizer(speech2)
tokens_nix=myTokenizer(speech3)
```

```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
wordcloud=WordCloud(stopwords=STOPWORDS,
                    background_color='black',
                    width=3000,
                    height=2500,
                    random_state=42
                   ).generate(speech1)
plt.figure(figsize=(10,7))
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
```
